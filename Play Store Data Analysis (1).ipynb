```python
def size_changes(x):
    if 'M' in x:
        val=float(x.replace('M',' '))
        val=val*1024
    elif 'k' in x:
        val=float(x.replace('k',' '))
    else:
        val=0
    return val
```

```python
df['Size']=df['Size'].apply(size_changes)
```

```python
df['Installs_n'].unique()
```

```python
def install(x):
    if '+' in x:
        val=x.replace('+','')
    if ',' in x:
        val=x.replace(',','')
    else:
        val=0
    return float(val)         
```

```python
df['Installs_n']=df['Installs_n'].apply(install)
```

```python
df['Installs_n']=df['Installs_n'].fillna(0)
```
